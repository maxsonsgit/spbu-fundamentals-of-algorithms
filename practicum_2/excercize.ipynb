```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.erdos_renyi_graph(n=10, p=0.2)
```

```python
def avg_neighbor_degree(G):
    neighbor_degree = []
    for n in G:
        neighbor_degree.append(len(list(G.neighbors(n))))
    return sum(neighbor_degree)/len(neighbor_degree)
```

```python
print(avg_neighbor_degree(G))
```

```python
def check_percolation(G):
    for n in G:
        if not(G.degree(n)):
            return 0
    return 1
```

```python
def progon(n, chance):
    data = []
    for graph in [nx.erdos_renyi_graph(n=n, p=chance) for _ in range(1000)]:
        data.append(check_percolation(graph))
    return sum(data)/1000
```

```python
x = np.linspace(0, 1, 20)
data = []
for el in x:
    data.append(progon(100, el))
y = data
    
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.plot(x, y, color="blue", linewidth=1.0, linestyle="--", marker="o")
ax.grid()
ax.set_xlabel(r'$p$', fontsize=12)
ax.set_ylabel(r'$Pperc$', fontsize=12)
plt.show()
```
